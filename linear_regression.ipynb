# Assignment Questions


##Q1. Import Required Libraries
Import all the necessary Python libraries for data handling and linear regression.

Example libraries: pandas, numpy, matplotlib, sklearn


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

## Q2. Load the Dataset
Load the dataset into a DataFrame using pandas.read_csv().
Display the first few rows using print.

```python
df = pd.read_csv('score.csv')
print(df.head())
```

## Q3. Prepare the Data
Separate the features (Hours) and target (Scores).
Use: X = df[['Hours']] y = df['Scores']


```python
x = df[['Hours']]
y = df['Scores']
```

## Q4. Train the Simple Linear Regression Model
Import LinearRegression from sklearn.linear_model.
Fit the model on all the data using .fit().

```python
model = LinearRegression()
model.fit(x,y)
```

## Q5. Make Predictions
Use the trained model to predict scores using .predict(X).
Display both actual and predicted scores side by side.

```python
predictions = model.predict(x)

comparison = pd.DataFrame({
    'Actual': y,
    'Predicted': predictions
})

print(comparison.head())
```

## Q6. Predict a Custom Value
Predict the score of a student who studies for 9.25 hours using the trained model


```python
predicted_score = model.predict(pd.DataFrame({'Hours': [9.25]}))
print(predicted_score)
```
